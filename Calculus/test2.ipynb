```python
def buildZ1(fList):
    x = np.linspace(-2, 2, 20)
    y = np.linspace(-2, 2, 20)

    X, Y = np.meshgrid(x, y)
    for f in fList:
        Z = f(X, Y)
    
        fig, ax = plt.subplots(figsize=(5,5))
        CS = ax.contour(X, Y, Z, cmap='RdGy')
        ax.xaxis.grid(True)
        ax.yaxis.grid(True)
        ax.clabel(CS, inline=1, fontsize=10)
        ax.set_title(str(f))
        plt.show()
```

```python
def build3D(fList):
    x = np.linspace(-2, 2, 20)
    y = np.linspace(-2, 2, 20)

    X, Y = np.meshgrid(x, y)
    for f in fList:
        Z = f(X, Y)
    
        fig, ax = plt.subplots(figsize=(5,5))
        ax = plt.axes(projection='3d')
        ax.contour3D(X, Y, Z, 50, cmap='binary')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        plt.xticks(rotation=90)
        plt.yticks(rotation=90)
        ax.set_title(str(f))
        plt.show()
```

```python
def f1(a, b):
    # return (a**2 + b**2) + 0.5*a -0.25*(a**2+b**2)**2
    # return (a**2 + b**2)**0.5
    return np.log(1 - a**2 - b)
def f2(a, b):
    return (a**3 + b**3)
    # return np.sin(a + b)
    # return np.sin(a - 3*b)
    return a**2 +b**2 -(a**2)*(b**2)
fList1 = [f2]
build3D(fList1)
buildZ1(fList1)
```

```python
from matplotlib.ticker import LinearLocator
#from matplotlib import cm

def draw_3d(X, Y, fct, zmin='min', zmax='max'):
    X, Y = np.meshgrid(X, Y)
    Z = fct(X, Y)
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    surf = ax.plot_surface(X, Y, Z, cmap='nipy_spectral', linewidth=0, antialiased=False)
    
    z1 = Z.min() if (zmin == 'min') else zmin
    z2 = Z.max() if (zmax == 'max') else zmax
    ax.set_zlim(z1, z2)
    ax.zaxis.set_major_locator(LinearLocator(10))
    
    ax.zaxis.set_major_formatter('{x:.02f}')
    
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()
    
```
