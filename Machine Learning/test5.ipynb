```python
## init2
d = 5
m = 3 # number of movements: a
gamma = 0.5
def init2():
    # initalization to zeroes
    T = np.zeros((m, d, d))
    R = np.zeros((m, d, d))
    V = np.zeros(d)
    
    # assign value to T, R, V
    # for stay movement
    stay = 0
    left = 1
    right = 2
    # for stay movement
    for n in np.arange(d):
        T[stay,n,n] = 1/2
        if (n != 0) and (n != d-1):
            T[stay,n,n-1] = 1/4
            T[stay,n,n+1] = 1/4
    T[stay,0,1] = 1/2
    T[stay,d-1,d-2] = 1/2
    
    # for left movement
    for n in np.arange(1,d):
        T[left,n,n-1] = 1/3
        T[left,n,n] = 2/3
    T[left,0,0] = 1/2
    T[left,0,1] = 1/2
    
    # for right movement
    for n in np.arange(d-1):
        T[right,n,n+1] = 1/3
        T[right,n,n] = 2/3
    T[right,d-1,d-1] = 1/2
    T[right,d-1,d-2] = 1/2
    
    for n in np.arange(m):
        R[n, d-1, d-1] = 1
    # R[stay, d-1, d-1] = 1
    return T,R,V
```
