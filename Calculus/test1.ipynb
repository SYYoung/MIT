```python
def buildZ(fList):
    x = np.linspace(-2, 2, 20)
    y = np.linspace(-2, 2, 20)

    X, Y = np.meshgrid(x, y)
    for f in fList:
        Z = f(X, Y)
    
        plt.contour(X, Y, Z, 20, cmap = 'RdGy')
        plt.title(str(f))
        plt.show()
```

```python
def buildZ1(fList):
    x = np.linspace(-2, 2, 20)
    y = np.linspace(-2, 2, 20)

    X, Y = np.meshgrid(x, y)
    for f in fList:
        Z = f(X, Y)
    
        fig, ax = plt.subplots(figsize=(5,5))
        CS = ax.contour(X, Y, Z, cmap='RdGy')
        ax.xaxis.grid(True)
        ax.yaxis.grid(True)
        ax.clabel(CS, inline=1, fontsize=10)
        ax.set_title(str(f))
        plt.show()
```

```python
from mpl_toolkits import mplot3d


def build3D(fList):
    x = np.linspace(-2, 2, 20)
    y = np.linspace(-2, 2, 20)

    X, Y = np.meshgrid(x, y)
    for f in fList:
        Z = f(X, Y)
    
        fig, ax = plt.subplots(figsize=(5,5))
        ax = plt.axes(projection='3d')
        ax.contour3D(X, Y, Z, 50, cmap='binary')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        plt.xticks(rotation=90)
        plt.yticks(rotation=90)
        ax.set_title(str(f))
        plt.show()
```

```python
def fun(a, b):
    return a*y - a**2
def gradx(a, b):
    return -2*a + b
def grady(a, b):
    return a
```

```python
def err(delta_x, delta_y):
    start_x, start_y = (2, 1)
    start = fun(start_x, start_y)
    pred = start + gradx(start_x, start_y) * delta_x + grady(start_y, start_y) * delta_y
    actual = fun(start_x+delta_x, start_y+delta_y)
    error = (actual - pred)/actual
    print('pred = ' + str(pred) + ', actual = ' + str(actual) + ', error = ' + str(error))
```

```python
err(0.1, 0.1)
```
